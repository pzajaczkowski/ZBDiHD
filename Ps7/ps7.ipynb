Wczytaj zbiór danych Titanic.csv.


```python
import pandas as pd

df = pd.read_csv("Titanic.csv")
```

Wyświetl 10 losowych wierszy z ramki danych.


```python
df.sample(10)
```

Wyświetl liczbę wierszy w ramce danych.


```python
df.shape[0]
```

Wyświetl szczegółowe informacje o ramce danych.


```python
df.info()
```

Zastąp brakujące wartości średnią poszczególnych kolumn.


```python
df.fillna(df.mean(numeric_only=True))
```

Posortuj zbiór danych po wieku malejąco.


```python
df.sort_values(by="Age", ascending=False)
```

Wypisz jednolinijkowym poleceniem wiersze wszystkich pasażerów, którzy:

- Byli mężczyznami.
- Przeżyli.
- Mieli najwyżej 40 lat.
- Przeżyli lub mieli poniżej 10 lat.
- Mieli Miss w imieniu.
- Wskaż najmłodszego pasażera.
- Wskaż najmłodszą pasażerkę.
- Wskaż najstarszego pasażera.
- Wskaż najstarszą pasażerkę.


- Byli mężczyznami


```python
df[df["Sex"] == "male"]
```

- Przeżyli


```python
df[df["Survived"] == 1]
```

- Mieli co najwyżej 40 lat


```python
df[df["Age"] <= 40]
```

- Przeżyli lub mieli poniżej 10 lat.


```python
df[(df["Survived"] == 1) | (df["Age"] <= 10)]
```

- Mieli Miss w imieniu.


```python
df[df["Name"].str.contains("Miss")]
```

- Wskaż najmłodszego pasażera.
- Wskaż najmłodszą pasażerkę.
- Wskaż najstarszego pasażera.
- Wskaż najstarszą pasażerkę.


```python
df[df["Sex"] == "male"].sort_values("Age").iloc[0]
```

```python
df[df["Sex"] == "female"].sort_values("Age").iloc[0]
```

```python
df[df["Sex"] == "male"].sort_values("Age", ascending=False).iloc[0]
```

```python
df[df["Sex"] == "female"].sort_values("Age", ascending=False).iloc[0]
```

Zapisz posortowany zbiór danych bez pustych wartości i indeksu do pliku Processed.csv


```python
df.sort_values("Age", ascending=False).dropna().to_csv("Processed.csv", index=False)
```
